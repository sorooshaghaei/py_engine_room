# 04 Logging, Config, Reproducibility\n\nUse config + logs + artifacts together so runs can be audited and repeated.

```python
import logging\nfrom pathlib import Path\n\nconfig = {"run_name": "demo", "root_dir": "."}\nlogging.basicConfig(level=logging.INFO, format="%(levelname)s:%(message)s")\n\nrun_dir = Path(config["root_dir"]) / "outputs" / config["run_name"]\nrun_dir.mkdir(parents=True, exist_ok=True)\nartifact = run_dir / "trace.txt"\nartifact.write_text("reproducible run\n", encoding="utf-8")\nlogging.info("artifact=%s", artifact)\nartifact.exists()
```

## Checklist\n- [ ] Config file committed (without secrets)\n- [ ] Outputs written under run-specific directory\n- [ ] Logs include step names\n\n## Common mistakes\n- Using `print` for production tracing\n- Overwriting artifacts without run names
