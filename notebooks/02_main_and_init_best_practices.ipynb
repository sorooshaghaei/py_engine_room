# 02 main.py And __init__.py Best Practices\n\n`main.py` should be thin: parse args -> load config -> call run() -> exit code.\n`__init__.py` should be minimal: public API + version only.

```python
thin_main = '''\ndef main() -> int:\n    args = parse_args()\n    config = load_config(args.config)\n    return run(config)\n'''\n\nminimal_init = '''\nfrom .pipeline import run_pipeline\n__version__ = "0.1.0"\n__all__ = ["run_pipeline", "__version__"]\n'''\n\nprint(thin_main)\nprint(minimal_init)
```
